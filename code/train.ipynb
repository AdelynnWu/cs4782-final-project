### Copy subject images from Google Drive to the project directory

The previous error indicated that no subject images were found. This step will copy your images from your Google Drive (`/content/drive/MyDrive/`) to the expected location (`/content/cs4782-final-project/cs4782-final-project/data/sompong_images`).

**Note**: This code will look for files starting with `IMG_` and ending with `.jpeg` or `.jpg` in your Google Drive. If your images have different names or formats, you might need to adjust the patterns below.

```python
import os
import shutil
from glob import glob

subject_target_dir = "/content/cs4782-final-project/cs4782-final-project/data/sompong_images"
os.makedirs(subject_target_dir, exist_ok=True)
print(f"Ensured directory exists: {subject_target_dir}")

# Path to the source images in MyDrive
source_image_dir = "/content/drive/MyDrive"

# List image files in MyDrive with supported extensions
image_patterns = [
    os.path.join(source_image_dir, 'IMG_*.jpeg'),
    os.path.join(source_image_dir, 'IMG_*.jpg')
]

source_images_paths = []
for pattern in image_patterns:
    source_images_paths.extend(glob(pattern))

print(f"Found {len(source_images_paths)} potential subject images in {source_image_dir} to copy.")

if not source_images_paths:
    print("Warning: No suitable image files (IMG_*.jpeg, IMG_*.jpg) found in MyDrive.")
    print("Please ensure your subject images are in /content/drive/MyDrive/ and have names like 'IMG_xxxx.jpeg' or 'IMG_xxxx.jpg'")
else:
    for src_path in source_images_paths:
        img_name = os.path.basename(src_path)
        dst_path = os.path.join(subject_target_dir, img_name)
        try:
            shutil.copy(src_path, dst_path)
            print(f"Copied {img_name} to {subject_target_dir}")
        except Exception as e:
            print(f"Error copying {img_name} from {src_path}: {e}")

print("\nImage copying attempt complete. Please verify the images are now in the target directory.")
print(f"You can try listing the contents of the target directory to confirm: `!ls -l {subject_target_dir}`")
```
